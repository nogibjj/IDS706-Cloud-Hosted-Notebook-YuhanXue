```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn.objects as so
import matplotlib.pyplot as plt

# Filter out FutureWarnings
import warnings

pd.set_option("mode.copy_on_write", True)

df = pd.read_csv(
    "https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv"
)
df.head()
```

```python


warnings.simplefilter(action="ignore", category=FutureWarning)

fig, ax = plt.subplots(figsize=(10, 7))
p = (
    so.Plot(df, x="Age", y="BMI")
    .add(so.Bar())
    .label(
        x="Age",
        y="BMI",
        title="Age and BMI",
    )
)
p.on(ax).plot()
plt.show()
```

```python
filtered_df = df[
    (df["Age"] < 30)
    | (df["Age"] > 50)
]
```

```python
fig, ax = plt.subplots(figsize=(10, 7))
p = (
    so.Plot(filtered_df, x="Age", y="BMI")
    .add(so.Bar())
    .label(
        x="Age",
        y="BMI",
        title="Age and BMI",
    )
)
p.on(ax).plot()
plt.show()
```
